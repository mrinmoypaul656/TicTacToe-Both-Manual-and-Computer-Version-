```python
#Implementation of Two Player Tic-Tac-Toe game in Python by Mrinmoy Paul
'''Firstly,
In every round of selecting position for playing game then a user can either
manually select the position by sellecting the position from 1-9 which is empty or
automatically he/she can choose the mark option by typing mark, the automatically the
a random position will be assigned to that user based on the availability of the position.

Secondly, if user1 and User2 is playing, and suddenly user1 wants to quit the game in between
hence he/she can quit by typing quit and then the opponent player wins directly.

Thirdly, after each game,a user can play the game again if he wants to play.'''

import random

''' We will make the board using dictionary 
    in which keys will be the location(i.e : top-left,mid-right,etc.)
    and initialliy it's values will be empty space and then after every move 
    we will change the value according to player's choice of move. '''

theBoard = {'1': ' ' , '2': ' ' , '3': ' ' ,
            '4': ' ' , '5': ' ' , '6': ' ' ,
            '7': ' ' , '8': ' ' , '9': ' ' }

board_keys = []

for key in theBoard:
    board_keys.append(key)

''' We will have to print the updated board after every move in the game and 
    thus we will make a function in which we'll define the printBoard function
    so that we can easily print the board everytime by calling this function. '''

def printBoard(board):
  print('\n')
  print(' ' + board['1'] + ' | ' + board['2'] + ' | ' + board['3'])
  print('-----------')    
  print(' ' + board['4'] + ' | ' + board['5'] + ' | ' + board['6'])
  print('-----------')
  print(' ' + board['7'] + ' | ' + board['8'] + ' | ' + board['9'])
  print('\n')

def takeInput():
  run = True
  while run:
    move = input()
    if (move == "quit" or move == "mark"):
      return move
    try:
      move = int(move)
      if move > 0 and move < 10:
        return str(move)
      else:
        print('Please type a number within the range!')
    except:
      print('Please type a number!')

def getWinner(turn, player1, player2):
  print("\nGame Over.\n")
  if(turn=='X'):           
    print(" **** " + player1 + " won. ****")
  if(turn=='O'):           
    print(" **** " + player2 + " won. ****")


# Now we'll write the main function which has all the gameplay functionality.
def game():
  player1 = input("Who's Player 1?\n")
  player2 = input("\nWho's Player 2?\n")

  turn = 'X'
  count = 0
  filled = ['0']

  for i in range(10):
    printBoard(theBoard)
    if(turn == 'X'):
      print("It's your turn, " + player1 + ".Move to which place?")
    if(turn == 'O'):
      print("It's your turn, " + player2 + ".Move to which place?")

    move = takeInput()

    if(turn=='X' and move=="quit"):
      print("\nGame Over.\n")
      print(" **** " + player2 + " won. ****")
      break

    if(turn=='O' and move=="quit"):
      print("\nGame Over.\n")
      print(" **** " + player1 + " won. ****")
      break


    if(move=="mark"):
      print("\nThe mode is Automatic, Assigning a random move.")
      move = '0'
      while(move in filled):
        move = str(random.randint(1,9))
    

    if theBoard[move] == ' ':
      theBoard[move] = turn
      count += 1
    else:
      print("\nThat place is already filled.\nMove to which place?")
      continue

    filled.append(move)

      # Now we will check if player X or O has won,for every move after 5 moves. 
    if count >= 5:
      if theBoard['7'] == theBoard['8'] == theBoard['9'] != ' ': # across the bottom
        printBoard(theBoard)
        getWinner(turn, player1, player2)
        break 
      elif theBoard['4'] == theBoard['5'] == theBoard['6'] != ' ': # across the middle
        printBoard(theBoard)
        getWinner(turn, player1, player2)
        break 
      elif theBoard['1'] == theBoard['2'] == theBoard['3'] != ' ': # across the top
        printBoard(theBoard)
        getWinner(turn, player1, player2)
        break 
      elif theBoard['1'] == theBoard['4'] == theBoard['7'] != ' ': # down the left side
        printBoard(theBoard)
        getWinner(turn, player1, player2)
        break 
      elif theBoard['2'] == theBoard['5'] == theBoard['8'] != ' ': # down the middle
        printBoard(theBoard)
        getWinner(turn, player1, player2)
        break 
      elif theBoard['3'] == theBoard['6'] == theBoard['9'] != ' ': # down the right side
        printBoard(theBoard)
        getWinner(turn, player1, player2)
        break 
      elif theBoard['7'] == theBoard['5'] == theBoard['3'] != ' ': # diagonal
        printBoard(theBoard)
        getWinner(turn, player1, player2)
        break 
      elif theBoard['1'] == theBoard['5'] == theBoard['9'] != ' ': # diagonal
        printBoard(theBoard)
        getWinner(turn, player1, player2)
        break 

    # If neither X nor O wins and the board is full, we'll declare the result as 'tie'.
    if count == 9:
      print("\nGame Over.\n")                
      print("It's a Tie!!")

    # Now we have to change the player after every move.
    if turn =='X':
      turn = 'O'
    else:
      turn = 'X'        
  
  # Now we will ask if player wants to restart the game or not.
  restart = input("\nDo want to play Again?(y/n)")

  if restart == "y" or restart == "Y" or restart == "yes" or restart == "YES" or restart == "Yes":  
    for key in board_keys:
      theBoard[key] = " "
    game()
  
  elif restart == "n" or restart == "N" or restart == "NO" or restart == "No" or restart == "no":
    print("\nThank You for Playing!")
  
  else:
    print("\nExecute the Program Again to Restart!")
    


if __name__ == "__main__":
    game()
```
